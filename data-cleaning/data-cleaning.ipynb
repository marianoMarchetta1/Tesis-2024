# Imports and dataset load

```python
import pandas as pd
from IPython.display import display
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('/Users/mmarchetta/Desktop/Tesis-2024/data-visualization/final_dataset_normalized.csv')
display(dataset.head())
```

```python
print(dataset['Tweets_Utilizados_referentes'].describe())
```

```python
dataset.drop(['Quote_asset_volume_BTCUSDT','ADX','CCI', 'MFI', 'Cantidad_pos_referentes', 'Cantidad_neg_referentes', 'Cantidad_neu_referentes', 'Compound total referentes', 'Sentimiento', 'Sentimiento_coin', 'Sentimiento_referentes'], axis=1, inplace=True)
dataset.to_csv('final_dataset_first_cleaning.csv', index=False, float_format='%.8f')
display(dataset.head())
```

```python
display(len(dataset.columns))
```

```python
# Eliminar las columnas de fecha y otras no numéricas
dataset2 = dataset.drop(columns=["Open_time"])

# Convertir cualquier variable no numérica en una variable numérica utilizando LabelEncoder
label_encoder = LabelEncoder()
for column in dataset2.columns:
    if dataset2[column].dtype == 'object':
        dataset2[column] = label_encoder.fit_transform(dataset2[column])

# Separar las características y la variable objetivo
X = dataset2.drop(columns=["Tendencia"])  # Características
y = dataset2["Tendencia"]  # Variable objetivo

# Seleccionar las 10 mejores características
k_best_features = SelectKBest(score_func=f_classif, k=35)
X_best = k_best_features.fit_transform(X, y)

# Obtener los índices de las características seleccionadas
selected_features_indices = k_best_features.get_support(indices=True)

# Obtener los nombres de las características seleccionadas
selected_features = X.columns[selected_features_indices]

# Mostrar las características seleccionadas
print("Características seleccionadas:")
print(selected_features)

dataset2 = dataset2[selected_features]
dataset2.insert(0, "Open_time", dataset["Open_time"])
dataset2["Tendencia"] = dataset["Tendencia"]

dataset2.to_csv('final_dataset_cleaned_kbest.csv', index=False, float_format='%.8f')
```

```python
# Eliminar las columnas "Open_time" y "Tendencia" del conjunto de datos
X_train_full = dataset.drop(columns=["Open_time", "Tendencia"])

# Obtener las etiquetas de la variable objetivo
y_train_full = dataset["Tendencia"]

# Entrenar un modelo de Random Forest utilizando todas las características
rf_model = RandomForestClassifier(
    n_estimators=500,  # Usar 500 árboles en el bosque
    criterion='gini',  # Criterio para dividir nodos: índice Gini
    max_depth=None,  # Sin límite en la profundidad de los árboles
    min_samples_split=2,  # Dividir un nodo si tiene al menos 2 muestras
    min_samples_leaf=1,  # Requerir al menos 1 muestra en cada hoja
    min_weight_fraction_leaf=0.0,  # Sin peso mínimo de muestra en cada hoja
    max_features='sqrt',  # Número máximo de características a considerar al dividir nodos: raíz cuadrada del número de características
    max_leaf_nodes=None,  # Sin límite en el número máximo de nodos hoja
    min_impurity_decrease=0.0,  # Un nodo se dividirá si la división induce una disminución de la impureza mayor o igual a este valor
    bootstrap=True,  # Realizar el muestreo bootstrap (remuestreo con reemplazo)
    oob_score=False,  # No calcular la puntuación fuera de la bolsa
    n_jobs=None,  # Usar todos los núcleos de CPU disponibles para el entrenamiento
    random_state=42,  # Semilla aleatoria para la reproducibilidad
    verbose=0,  # Sin salida detallada durante el entrenamiento
    warm_start=False,  # Comenzar desde el último entrenamiento
    class_weight=None,  # Sin ponderación de clases
    ccp_alpha=0.0,  # Parámetro de complejidad del costo mínimo para la poda del costo-complejidad
    max_samples=None  # Número máximo de muestras a usar para cada árbol
)

rf_model.fit(X_train_full, y_train_full)

# Utilizar SelectFromModel para seleccionar características basadas en la importancia del modelo
selector = SelectFromModel(rf_model, prefit=True, threshold=0.02)

# Obtener las características seleccionadas
selected_features = X_train_full.columns[selector.get_support()]

# Filtrar las características seleccionadas
selected_data = dataset[selected_features].copy()  # Crear una copia para evitar la advertencia

# Añadir la variable "Tendencia" al DataFrame seleccionado
selected_data["Tendencia"] = y_train_full

# Guardar el conjunto de datos con las características seleccionadas y la variable Tendencia
selected_data.insert(0, "Open_time", dataset["Open_time"])
selected_data.to_csv('final_dataset_cleaned_random_forest_feature_selector.csv', index=False)

print(len(selected_features))
```
